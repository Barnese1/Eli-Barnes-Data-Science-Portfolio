```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins').dropna()
df.head()
```

```python
sns.histplot(
    data=df,
    x="bill_length_mm",
    y="flipper_length_mm",
    bins=12,
    cmap="viridis",
    cbar=True  
)
plt.savefig("hist_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sns.kdeplot(
    data=df,
    x="bill_length_mm",
    y="flipper_length_mm",
    fill=True,          # Fills the contours
    cmap="viridis",     # Color map
    thresh=0.05,        # Removes very low-density noise
    levels=100          # Smoothness/detail level
)
plt.savefig("kde_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```
